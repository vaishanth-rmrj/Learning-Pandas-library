```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading the dataset
df = pd.read_csv('data/literacy_rates_data.csv')
df.head()
```

1. Filter by specific country

```python
df = df.iloc[0:150]
print(df['Country'].value_counts())
fig = plt.figure(figsize = (10, 5))
plt.bar(df['Country'].value_counts().keys(), df['Country'].value_counts().values, width=0.4)
plt.show()
```

```python
filter = (df['Country'] == 'Afghanistan')
df[filter].head()
```
